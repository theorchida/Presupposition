```python
import pandas as pd
```

```python
shots = 6
```

```python
df = pd.read_excel("nope_data/gpt_response.xlsx")
df
```

```python
df_copy = df.loc[lambda df: (df["gpt_response"] == "E") | (df["gpt_response"] == "N")]
# print(df.loc[lambda df: df["gpt_response"] in ["E", "N"]])
df_copy
```

```python
# number of discarded responses
discarded = len(df) - shots - len(df_copy)
discarded
```

```python
def get_confusion_matrix(shots, df_copy):
    tp = 0
    tn = 0
    fp = 0
    fn = 0

    for i in range(len(df_copy)):
        label = df_copy.iloc[i]['label']
        gpt_response = df_copy.iloc[i]['gpt_response']
        
        if (label == 'E') and (gpt_response == 'E'):
            tp += 1
        elif (label in ['N', 'C']) and (gpt_response == 'N'):
            tn += 1
        elif (label == 'E') and (gpt_response == 'N'):
            fn += 1
        else:
            fp += 1
    
    return tp, tn, fp, fn

```

```python
def get_metrics(tp, tn, fp, fn):
    accuracy = (tp+tn) / (tp+tn+fp+fn)
    precision = tp / (tp+fp)
    recall = tp / (tp+fn)
    f1_score = 2*precision*recall / (precision+recall)
    return accuracy, precision, recall, f1_score
```

```python
tp, tn, fp, fn = get_confusion_matrix(shots, df_copy)
print(tp+tn+fp+fn)
```

```python
accuracy, precision, recall, f1_score = get_metrics(tp, tn, fp, fn)
print(accuracy, precision, recall, f1_score)
```
